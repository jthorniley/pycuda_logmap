```python
import pycuda.autoinit
import pycuda.driver as drv
import numpy

from pycuda.compiler import SourceModule


HISTOGRAM_SIZE = 4000
BLOCK_SIZE=512
HIST_BOTTOM=-0.65
HIST_TOP=0.65
HIST_RANGE = HIST_TOP-HIST_BOTTOM

COUPLING_START = 0.0
COUPLING_END = 0.35

mod = SourceModule("""



__global__ void generate_histogram(uint *d_histogram, uint discard, uint nsteps,
  float *random_starting_points, float coupling_start, float coupling_end )
{
    //A block performs repeated 
    const uint hist_size = %d;
    const uint shmem_size = hist_size + 2;
    const float hist_bottom = %f;
    const float hist_range = %f;
    __shared__ uint hist[shmem_size];
   
    for ( uint i = 0; i <= shmem_size/blockDim.x; ++i ) {
        uint idx = threadIdx.x + i * blockDim.x;
        if ( idx < shmem_size) {
            hist[idx] = 0;
        }
    }
    __syncthreads();
    //Coupling is fixed by blockIdx.x
   
    float coupling = coupling_start + (coupling_end-coupling_start)/gridDim.x * blockIdx.x;
   
    float x = random_starting_points[2*threadIdx.x];
    float y = random_starting_points[2*threadIdx.x+1];
    float x1, y1;
    float bxy = 0.02;
   
    for (uint i = 0; i < discard; ++i ) {
        x1 = (1-bxy) * x + bxy * y;
        y1 = (1-coupling) * y + coupling * x;
        x = x1*3.7*(1.0-x1);
        y = y1*3.7*(1.0-y1);
    
        
    }

    __syncthreads();
    
    for (uint i = 0; i < nsteps; ++i ) {
        x1 = (1-bxy) * x + bxy * y;
        y1 = (1-coupling) * y + coupling * x;
        x = x1*3.7*(1.0-x1);
        y = y1*3.7*(1.0-y1);
    
    
        float diff = x-y;
        
        int point = (diff-hist_bottom)/hist_range * hist_size + 1;
        point = point > hist_size ? hist_size+1 : point;
        point = point < 1 ? 0 : point;
        
        atomicAdd( &hist[point], 1);
    }
    
    __syncthreads();
    
    for ( uint i = 0; i <= hist_size/blockDim.x; ++i ) {
        uint idx = threadIdx.x + i * blockDim.x;
        if ( idx < hist_size ) {
            d_histogram[hist_size*blockIdx.x + idx] += hist[idx+1];
        }
    }
    
}
""" % (HISTOGRAM_SIZE, HIST_BOTTOM, HIST_RANGE))

generate_histogram = mod.get_function("generate_histogram")
```

```python
from pycuda import gpuarray
```

```python
NCOUPLING=4000

d_result = gpuarray.zeros((NCOUPLING,HISTOGRAM_SIZE),dtype=uint32)
                        
```

```python
for i in range(50):
    start = rand(BLOCK_SIZE,2).astype(float32)
    d_start = gpuarray.to_gpu(start)
    generate_histogram(
        d_result.gpudata, uint32(500), uint32(500), d_start.gpudata, float32(COUPLING_START), float32(COUPLING_END),
        block=(BLOCK_SIZE,1,1), grid=(NCOUPLING,1))
```

```python
result = d_result.get()
```

```python
hist_data = result.astype(float)
for i in range(result.shape[0]):
    row = hist_data[i,:]
    hist_data[i,:] /= max(row)
```

```python
from scipy.ndimage.filters import gaussian_filter
coupling_range = arange(COUPLING_START,COUPLING_END,(COUPLING_END-COUPLING_START)/NCOUPLING)
figure(dpi=600,figsize=(float(NCOUPLING)/500.0,float(HISTOGRAM_SIZE)/500.0))
imdata = (gaussian_filter((hist_data)/2,(0.5,0.0)))

imshow(imdata.T,origin="lower",extent=(coupling_range[0],coupling_range[-1],HIST_BOTTOM,HIST_TOP),aspect=0.25,interpolation="bicubic",cmap=cm.bone)
gca().tick_params(top="off",right="off",direction="out")
gca().spines["top"].set_color("none")
gca().spines["right"].set_color("none")
axis("off")
extent = gca().get_window_extent().transformed(gcf().dpi_scale_trans.inverted())
text(0.2,-0.5,"Jelly!",color="white",name="Impact",size=50)
savefig("bidirectional-map-plot-jelly.png",dpi=600,bbox_inches=extent)
```
